```python
import pandas as pd
from pathlib import Path

root_path = Path("..")
preprocessed_train_data = root_path / "data" / "preprocessed_train_data"
preprocessed_val_data = root_path / "data" / "preprocessed_val_data"

df_train = pd.read_csv(preprocessed_train_data / "train_data.csv")
df_val = pd.read_csv(preprocessed_val_data / "val_data.csv")
```

```python
df_train
```

```python
df_val
```

```python
personality_labels = [
    "Honesty-Humility",
    "Extraversion",
    "Agreeableness",
    "Conscientiousness",
]
performance_labels = [
    "Integrity",
    "Collegiality",
    "Social_versatility",
    "Development_orientation",
    "Hireability",
]
```

```python
train_columns = [
    col
    for col in df_train.columns
    if col not in personality_labels
    and col not in performance_labels
    and col not in ["id"]
]

train_columns
```

```python
X_train = df_train[train_columns]
X_val = df_val[train_columns]

print(X_train.shape, X_val.shape)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

metrics = {}
for label in personality_labels:
    y_train = df_train[label]

    model = Ridge(random_state=42)
    model.fit(X_train, y_train)

    y_true = df_val[label]
    y_pred = model.predict(X_val)

    metrics[label] = {
        "mse": mean_squared_error(y_true, y_pred),
        "r2": r2_score(y_true, y_pred),
    }
```

```python
from rich import print as pprint

pprint(metrics)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

metrics = {}
for label in performance_labels:
    y_train = df_train[label]

    model = Ridge(random_state=42)
    model.fit(X_train, y_train)

    y_true = df_val[label]
    y_pred = model.predict(X_val)

    metrics[label] = {
        "mse": mean_squared_error(y_true, y_pred),
        "r2": r2_score(y_true, y_pred),
    }
```
